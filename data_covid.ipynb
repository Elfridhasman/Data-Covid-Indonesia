```python
import json
```

```python
with open("kawalcovid.json") as f:
    data = json.load(f)
```

```python
import pandas as pd
df = pd.DataFrame(columns=["FID","Kode_Provi","Provinsi","Kasus_Posi","Kasus_Semb","Kasus_Meni"])
```

```python
for i in range(0,len(data)):
    currentItem = data[i]
    df.loc[i] = [data[i]["attributes"]["FID"],data[i]["attributes"]["Kode_Provi"],data[i]["attributes"]["Provinsi"],data[i]["attributes"]["Kasus_Posi"],data[i]["attributes"]["Kasus_Semb"],data[i]["attributes"]["Kasus_Meni"]]
```

```python
df
```

```python
df.to_csv("data_covid.csv")
```
